```python
from collections import deque
```

```python
graph={
    'A':['B','C'],
    'B':['A','D','E'],
    'C':['A','F'],
    'D':['B'],
    'E':['B','F'],
    'F':['C','E']
}

```

```python
def BFS(graph,start):
    visited={start}
    queue=deque([start])

    while queue:
        node=queue.popleft()
        print(node,end=" ")
        for i in graph[node]:
            if i not in visited:
                visited.add(i)
                queue.append(i)
```

```python
BFS(graph,"A")
```
