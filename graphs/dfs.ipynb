```python
from collections import deque
```

```python
graph={
    'A':['B','C'],
    'B':['A','D','E'],
    'C':['A','F'],
    'D':['B'],
    'E':['B','F'],
    'F':['C','E']
}
```

```python
def DFS(graph,node,visited=None):
    if visited is None:
        visited=set()
    if node not in visited:
        print(node,end=" ")
        visited.add(node)
        for i in graph[node]:
            DFS(graph,i,visited)
    return visited
```

```python
result=DFS(graph,'A')
```
