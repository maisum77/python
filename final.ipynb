BFS:

```python
from collections import deque

Graph={
    'A':['B','C'],
    'B':['A','D','E'],
    'C':['A','F'],
    'D':['B','G'],
    'E':['B','G'],
    'F':['C','G'],
    'G':['D','E','F']
}

def BFS(Graph,start,goal):
    visited={start}
    Queue=deque([[start]])
    while Queue:
        path=Queue.popleft()
        node=path[-1]
        if node==goal:
            return path
        for i in Graph[node]:
            if i not in visited:
                visited.add(i)
                new_path=path+[i]
                Queue.append(new_path)
    return None

print(BFS(Graph,'A','G'))
```

```python
from collections import deque

def Is_safe(state): #state is like array index from 0to3 containing state of a thing currently
    F=state[0]
    G=state[1]
    W=state[2]
    C=state[3]
    if G==C and F!=G:
        return False
    elif W==G and F!=W :
        return False
    else:
        return True

def opposite(f):
    if f == "R":
        return "L"
    else:
        return "R"


def Safe_move(state):
    F=state[0]
    G=state[1]
    W=state[2]
    C=state[3]
    
    safe=[]

    new_state=[opposite(F),G,W,C]
    if (Is_safe(new_state)):
        safe.append(new_state)
    if F==G:
        new_state=[opposite(F),opposite(G),W,C]
        if (Is_safe(new_state)):
                safe.append(new_state)
    if F==W:
        new_state=[opposite(F),G,opposite(W),C]
        if (Is_safe(new_state)):
            safe.append(new_state)
    if F==C:
        new_state=[opposite(F),G,W,opposite(C)]
        if (Is_safe(new_state)):
            safe.append(new_state)
    
    return safe

def solving(initial_state):
    goal_state = ['R', 'R', 'R', 'R']
    queue = deque([ [initial_state] ])
    visited = { tuple(initial_state) }
    while queue:
        current_path = queue.popleft()
        current_state = current_path[-1]
        if current_state == goal_state:
            print(" Solution found!")
            return current_path  
        for next_state in Safe_move(current_state):
            if tuple(next_state) not in visited:
                visited.add(tuple(next_state))
                new_path = current_path + [next_state]
                queue.append(new_path)

  
    print("No solution found.")
    return None


def print_solution(path):
    if path is None:
        return
    
    print("\n--- (Farmer, Goat, Wolf, Cabbage) ---")
    for i, step in enumerate(path):
        print(f"Step {i}: {step}")


state = ['L', 'L', 'L', 'L']
solution_path = solving(state)
print_solution(solution_path)
```

```python
from collections import deque

Graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': ['G'],
    'E': ['G'],
    'F': ['G']
}

def shortest_path_bfs(graph, start, goal):
    # Queue stores tuples: (current_node, path_to_current_node)
    Queue = deque([(start, [start])])
    Visited = {start}

    while Queue:
        # Dequeue the current node and its path
        (current, path) = Queue.popleft()

        # Check if the goal is reached
        if current == goal:
            return path # Found the shortest path

        # Explore neighbors
        for neighbor in graph.get(current, []):
            if neighbor not in Visited:
                Visited.add(neighbor)
                # Create the new path by extending the current path
                new_path = path + [neighbor]
                # Enqueue the neighbor and the new path
                Queue.append((neighbor, new_path))
    
    return None # Goal not reachable

# Example Execution
shortest_path = shortest_path_bfs(Graph, 'A', 'G')
print(f"Shortest Path from A to G: {shortest_path}")
print(f"Minimum Steps: {len(shortest_path) - 1 if shortest_path else 0}")
```
