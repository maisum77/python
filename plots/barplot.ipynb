```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
phool=sns.load_dataset("iris")
phool
```

```python
#linegraph
plt.figure(figsize=(4,4))
sns.barplot(x="species",y="sepal_width",data=phool)
plt.title("iris ka plot")
plt.show()
```

```python
ship=sns.load_dataset("titanic")
ship
```

```python
sns.barplot(x="sex",y="age",data=ship,)
plt.show()
```

```python
sns.barplot(x="sex",y="age",data=ship,ci=None)
plt.show()
```

```python
sns.barplot(x="sex",y="age",data=ship,errorbar=None,hue='alone')
plt.show()
```

```python
sns.barplot(x="class",y="fare",data=ship,hue='alone',estimator=np.mean)
plt.figure(figsize=(6,6))
plt.show()
```
