```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
phool=sns.load_dataset("iris")
phool
```

```python
#linegraph
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("iris ka plot")
sns.set_style(None)
plt.show()
```
