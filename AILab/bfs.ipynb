```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque
```

```python
def Is_safe(state):
    f,w,g,c=state
    if(w==g) and (f!=w):
        return False
    if(g==c) and (f!=g):
        return False
    return True
```

```python
def successor(state):
    f,w,g,c=state
    moves=[]
    for item in ['W','G','C',None]:
        new_state=[f,w,g,c]
        if item =='W' and f == w:
            new_state[0]='R'  if f == 'L' else 'L'
            new_state[1]=new_state[0]
        elif item =='G' and f==g:
            new_state[0]='R' if f=='L' else 'L'
            new_state[2]=new_state[0]
        elif item =='C' and f==c:
            new_state[0]='R' if f=='L' else 'L'
            new_state[3]=new_state[0]
        elif item is None:
            new_state[0]='R' if f=='L' else 'L'
        if Is_safe(new_state):
            moves.append(new_state)
    return moves



```

```python
def bfs_farmer():
    start=['L','L','L','L']
    goal=['R','R','R','R']
    queue=deque([start,[start]])
    visited=(start)
    while queue:
        state,path=queue.popleft()
        if state == goal:
            return path
        for next_state in successor(state):
            if next_state not in visited:
                visited.add(next_state)
                queue.append((next_state,path +[next_state]))
    result=bfs_farmer()
    for step in result:
        print(step)
```
