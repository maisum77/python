```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque
```

```python
G= nx.Graph([("A","B"),("A","C"),("B","D"),("B","E"),("C","F"),("E","F")])
def BFS(s,g):#s for start point and g for goal point
    q,v=deque([[s]]),set()
    while q:
        p=q.popleft()
        if p[-1] ==g: return p
        v.add(p[-1])
        # q.extend([p+[n]for n in G.neighbors(p[-1])])
        for n in G.neighbors(p[-1]):
            if n not in v:
                q.extend([p + [n]])

print("BFS Path: ", BFS("A","F") )
nx.draw(G,with_labels=True, node_color="lightblue")
plt.show()
```

```python

G = nx.Graph([("A", "B"), ("A", "C"),("B", "D"), ("B", "E"), ("C", "F"), ("D", "G"), ("H", "I"), ("G", "H"),("I", "J"), ("J", "F")])

def DFS(s, g): # s for start point and g for goal point
    
    stack, v = deque([[s]]), set()
    
    while stack: 
        p = stack.pop() 

        if p[-1] == g: 
            return p
        if p[-1] not in v:
            v.add(p[-1])
            for n in G.neighbors(p[-1]):
                stack.extend([p + [n]]) 


path = DFS("A", "H")
print("DFS Path: ", path) 
nx.draw(G, with_labels=True, node_color="lightblue")
plt.show()
```
