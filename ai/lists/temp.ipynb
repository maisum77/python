```python
from collections import deque
```

```python
Graph={
    'S':['A','B'],
    'A':['C','D','S'],
    'B':['E','S'],
    'C':['A'],
    'D':['G','A'],
    'E':['B','G'],
    'G':['D','E']
}
```

```python
def BFS(graph,start,goal):
    visited={start}
    Queue=deque([start])
    while Queue:
        node=Queue.popleft()
        print(node,end=" ")
        if (node==goal):
            return None
        for i in graph[node]:
            if i not in visited:
                Queue.append(i)
                visited.add(i)
```

```python
BFS(Graph,"A","G")
```
