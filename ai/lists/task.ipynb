```python
Graph={
    'S':['A','B'],
    'A':['S','C','D'],
    'B':['E'],
    'C':['A'],
    'D':['A','G'],
    'E':['B','G'],
}
```

```python
def DFS(graph,goal,start,visited=None,path=None):
    if visited is None:
        visited=set()
        path=[]
    if start not in visited:
        visited.add(start)
        path.append(start)
    if start == goal:
        return path
    for i in graph[start]:
        if i not in visited:
            a=DFS(graph,goal,i,visited,path)
            if a is not None:
                return a
```

```python
DFS(Graph,"G","A")
```
