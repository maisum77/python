```python
from collections import deque
```

```python
Graph={
    'A':['B','C'],
    'B':['A','D','E'],
    'C':['A','F'],
    'D':['B','G'],
    'E':['B','G'],
    'F':['C','G'],
    'G':['D','E','F']
}
```

```python
def BFS(Graph,start,goal):
    visited={start}
    Queue=deque([[start]])
    while Queue:
        path=Queue.popleft()
        node=path[-1]
        if node==goal:
            return path
        for i in Graph[node]:
            if i not in visited:
                visited.add(i)
                new_path=path+[i]
                Queue.append(new_path)
    return None
```

```python
print(BFS(Graph,'A','G'))
```
